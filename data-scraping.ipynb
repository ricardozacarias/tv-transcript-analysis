```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# First I'll try to scrape the links to the transcipts of each episode.
url = 'https://transcripts.fandom.com/wiki/South_Park'

# get the response and cook the soup
response = requests.get(url)
print('Response code:', response.status_code)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# all episode links are contained within <div> with a specific style
seasons = soup.find_all('div', {'style':'column-count:2'})
```

```python
# in seasons we get a list of every tag containing that find_all pattern
# in this case, it is a list of lists (each element is a section
links = []
for season in seasons:
    for episode in season.find_all('a'):
        links += ['https://transcripts.fandom.com' + episode.get('href')]      
        
```
